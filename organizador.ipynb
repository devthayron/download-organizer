# ORGANIZADOR_DOWNLOALD

### importações

```python
from pathlib import Path
import shutil
import json
from pprint import pprint
```

### Recebe o arquivo json com as configurações

```python
with open('config.json',encoding='utf-8') as arquivo:
    config = json.load(arquivo)
pprint(config)
```

### Organiza a pasta downloald definido em `config.json`

```python
CAMINHO_DOWNLOADS = Path.home() / 'Downloads'

# cria as pastas baseado no config.json
for pasta in config['regras']:
    (CAMINHO_DOWNLOADS/pasta).mkdir(exist_ok=True)

for arquivo in CAMINHO_DOWNLOADS.iterdir():

    # Ignora as pastas e arquivos ocultos
    arquivos_ocultos = arquivo.name.startswith('.')
    if not arquivo.is_file() or arquivos_ocultos:
        continue

    extensao = arquivo.suffix.lower().lstrip('.')
    
    for pasta, extensoes in config['regras'].items():
        if extensao in extensoes:
            destino_base = CAMINHO_DOWNLOADS / pasta / arquivo.name

            # Verificar duplicata antes de mover
            destino = destino_base
            i = 1
            while destino.exists():
                destino = destino_base.with_stem(f"{destino_base.stem}({i})")
                i += 1
            shutil.move(arquivo,destino)
            break
    else:
        print(f"[SEM REGRA] {arquivo.name}")
```

### Refaz e apaga as pasta voltando ao estado original

```python
for pasta in config['regras']:
    caminho = CAMINHO_DOWNLOADS/pasta
    for arquivo in caminho.iterdir():
        shutil.move(arquivo, CAMINHO_DOWNLOADS)
    # apaga a pasta (apenas se tiver vazio)
    caminho.rmdir()
```
